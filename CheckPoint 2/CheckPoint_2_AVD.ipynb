### **Outliers Values**

Pada tahap ini, kita akan melihat nilai ekstrem (outliers) pada kolom yang bertipe data numerik. Hal ini dilakukan agar mendapatkan insight yang akurat sehingga dapat memperkuat analisis yang akan dilakukan.

Akan tetapi, tidak semua kolom perlu ditangani. Kita tetap harus melihat dan memahami konteks dari setiap kolom yang ada pada dataset.

Dan terlihat kolom:

* Quantity

Dengan persentase 0%

* Sales

Dengan persentase 1%

* Profit

Dengan persentase 5.26%

```python
results = []

cols = df.select_dtypes(include=['float64', 'int64'])

for col in cols:
  q1 = df[col].quantile(0.25)
  q3 = df[col].quantile(0.75)
  iqr = q3 - q1
  lower_bound = q1 - 1.5*iqr
  upper_bound = q3 + 1.5*iqr
  outliers = df[(df[col] < lower_bound) | (df[col] > upper_bound)]
  percent_outliers = (len(outliers)/len(df))*100
  results.append({'Kolom': col, 'Persentase Outliers': percent_outliers})

# Dataframe dari list hasil
results_df = pd.DataFrame(results)
results_df.set_index('Kolom', inplace=True)
results_df = results_df.rename_axis(None, axis=0).rename_axis('Kolom', axis=1)

# Tampilkan dataframe
display(results_df)
```

Berdasarkan hasil perhitungan, diperoleh hasil sebagai berikut:

* Quantity

Memiliki persentase outlier sebesar 0%, yang menunjukkan tidak terdapat nilai ekstrem pada kolom ini.

* Sales

Memiliki persentase outlier sebesar 1%, yang tergolong sangat kecil sehingga masih dapat dianggap wajar dalam konteks variasi transaksi penjualan.

* Profit

Memiliki persentase outlier sebesar 5.26%, yang menunjukkan terdapat beberapa nilai keuntungan yang cukup ekstrem dibandingkan distribusi normalnya.

Dengan demikian, kolom Profit perlu diperhatikan lebih lanjut untuk memastikan apakah nilai ekstrem tersebut merupakan kejadian wajar dalam bisnis atau perlu dilakukan penanganan khusus.

# **Eksplorasi Data (EDA)**

## **Comparison/Perbandingan**

Aktivitas : Membandingkan total penjualan (Sales) pada setiap wilayah (Region).

Tujuan : Untuk mengetahui wilayah mana yang memberikan kontribusi penjualan terbesar serta membandingkan performa penjualan antar wilayah secara langsung.

Visualisasi : Bar Chart (Grafik Batang).

```python
penjualan_region = df.groupby('Region')['Sales'].sum().sort_values(ascending=False)

plt.figure(figsize=(10,6))
sns.barplot(x=penjualan_region.index,
            y=penjualan_region.values,
            palette='viridis')

plt.title('Total Penjualan Berdasarkan Region')
plt.xlabel('Region')
plt.ylabel('Total Penjualan')
plt.xticks(rotation=45)
plt.show()
```

Grafik ini membandingkan total penjualan dari setiap wilayah.

Wilayah West memiliki total penjualan paling tinggi dibanding wilayah lainnya. Ini berarti wilayah tersebut memberikan kontribusi terbesar terhadap pendapatan.

Wilayah East dan South memiliki total penjualan yang hampir sama dan berada di posisi tengah.

Sedangkan wilayah North memiliki total penjualan paling rendah dibanding wilayah lain, meskipun tetap memberikan kontribusi yang cukup besar.

## **Composition/Komposisi**

Aktivitas: Melihat komposisi penjualan berdasarkan Category.

Tujuan: Mengidentifikasi kategori dengan kontribusi penjualan terbesar dibandingkan kategori lainnya.

Visualisasi: Pie Chart.

```python
sales_by_category = df.groupby('Category')['Sales'].sum().sort_values(ascending=False)
sales_by_category_top3 = sales_by_category.head(3)

plt.figure(figsize=(10, 8))
sales_by_category_top3.plot(kind='pie', autopct='%1.1f%%', startangle=140, colors=plt.cm.Paired.colors)
plt.title('Proporsi 3 Kategori Produk Teratas Berdasarkan Total Penjualan')
plt.ylabel('')
plt.axis('equal')
plt.show()
```

Grafik ini menunjukkan persentase penjualan berdasarkan kategori produk.

Kategori Electronics memberikan kontribusi terbesar, yaitu sekitar 49,9% dari total penjualan. Artinya, hampir setengah dari seluruh penjualan berasal dari produk elektronik.

Di posisi kedua ada kategori Accessories, yang menyumbang sekitar 39,8% dari total penjualan.

Sedangkan kategori Office memberikan kontribusi paling kecil, yaitu sekitar 10,3% dari total penjualan.

## **Distribution/Distribusi**

Aktivitas : Menganalisis distribusi nilai transaksi pada kolom Sales.

Tujuan : Untuk mengetahui pola penyebaran nilai transaksi, melihat apakah mayoritas transaksi bernilai kecil atau besar, serta mengidentifikasi adanya kecenderungan distribusi (skewness) pada data penjualan.

Visualisasi : Histogram.

```python
plt.figure(figsize=(10, 6))
sns.histplot(df['Sales'], bins=20, kde=True)
plt.title('Distribusi Penjualan (Sales)')
plt.xlabel('Sales')
plt.ylabel('Frekuensi')
plt.show()
```

Distribusi Sales tidak berbentuk normal karena tidak simetris. Grafik menunjukkan pola right-skewed (positively skewed), dimana sebagian besar transaksi berada pada nilai rendah dan hanya sedikit transaksi dengan nilai sangat tinggi.

## **Relationship/Hubungan**

Aktivitas : Menganalisis hubungan antara nilai penjualan (Sales) dengan keuntungan (Profit).

Tujuan : Untuk mengetahui apakah peningkatan nilai penjualan berbanding lurus dengan peningkatan profit, serta mengidentifikasi kemungkinan adanya transaksi dengan penjualan tinggi namun menghasilkan profit rendah atau bahkan kerugian.

Visualisasi : Heatmap

```python
plt.figure(figsize=(8, 6))
sns.heatmap(data=df[['Sales', 'Profit']].corr(),
            annot=True,
            cmap='coolwarm',
            fmt='.2f')
plt.title('Korelasi antara Sales dan Profit')
plt.show()
```

Heatmap digunakan untuk melihat hubungan antar variabel dalam bentuk korelasi. Pada visualisasi ini, warna merah menunjukkan hubungan positif, sedangkan warna biru menunjukkan hubungan negatif.

Berdasarkan heatmap tersebut, nilai korelasi antara Sales dan Profit adalah 0,83.

Nilai 0,83 menunjukkan bahwa Sales dan Profit memiliki hubungan positif yang kuat. Artinya, ketika penjualan meningkat, keuntungan juga cenderung meningkat.

Warna merah pada bagian diagonal menunjukkan nilai 1,00 karena variabel dibandingkan dengan dirinya sendiri. Sementara itu, nilai 0,83 antara Sales dan Profit menunjukkan hubungan yang kuat dan searah.
